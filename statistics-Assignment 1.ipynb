# Problem 1

You have to create a function that receives a list of numbers like [23,14,13,56,78,98,12] and returns it's Interquartile range(IQR)

Note: You have to code the logic from scratch

```python
def interquartile(l):
    if(len(l)%2!=0):
       #odd
       
        l1=[]
        l2=[]
        l.sort()
        full=len(l)
        half=int(len(l)/2)
        for i in range(half):
            l1.append(l[i])
        for j in range(half+1,full):
            l2.append(l[j])
        a=l1[int(len(l1)/2)]
        b=l2[int(len(l2)/2)]
        result=(b-a)
        print("The interquartile range : ",result)
    else:
        #even
       
        l1=[]
        l2=[]
        l.sort()
        full=len(l)
        half=int(len(l)/2)+1
        for i in range(half-1):
            l1.append(l[i])
        for j in range(half-1,full):
            l2.append(l[j])
        a=l1[int(len(l1)/2)]
        b=l2[int(len(l2)/2)]
        result=(b-a)
        print("The interquartile range : ",result)
```

```python
l=[3,5,7,8,9,11,15,16,20,21]
interquartile(l)
```

```python
l=[1,2,5,6,7,9,12,15,18,19,27]
interquartile(l)
```

```python
l=[23,14,13,56,78,98,12]
interquartile(l)
```

# Problem 2

Plot a histogram for total team scores for all IPL games.

Note: Per match 2 teams will play hence total number scores would be = no. of matches * 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
M=pd.read_csv("matches.csv")
D=pd.read_csv("deliveries.csv")
```

```python
M.head()
```

```python
D.head(200)
```

```python
mega=M.merge(D,left_on='id',right_on='match_id')
mega.info()
```

```python
total_team_score_per_match=D.groupby('match_id').sum()['total_runs']
total_team_score_per_match
```

```python
plt.hist(total_team_score_per_match.values, bins = 50,color='red',edgecolor='k')
plt.title('Total Team score for all IPL games')
plt.show()
```

```python
score=D.groupby(['match_id','inning']).sum()['total_runs']
score
```

```python
plt.hist(score.values, bins = 50,color='red',edgecolor='k')
plt.title('Total Team score for all IPL games')
plt.show()
```

# Problem 3

Find the skewness of bowler economy data where the bowler has bowled a minimum of 300 balls.

Note : if you don't know about economy read this https://en.wikipedia.org/wiki/Economy_rate_(cricket)

```python
D
```

```python
bowlers_balls=D.groupby('bowler').count()['ball']
bowlers_balls=bowlers_balls[bowlers_balls>300]
over_bowled=bowlers_balls/6

bowlers_runs=D.groupby('bowler').sum()['batsman_runs']
economy_rate=bowlers_runs/over_bowled
economy_rate.dropna(inplace=True)
economy_rate.skew()
sns.kdeplot(economy_rate)
plt.title('The skewness of bowler economy rate')
plt.show()
```
